```python
import pandas as pd
import seaborn as sns

# Leer el archivo de correlaciones
app_corrs = pd.read_csv("../datasets/correlations/apps_correlations.csv")

# Crear tabla pivote con todas las métricas
table = app_corrs.pivot_table(
    index=['APPLICATION_NAME', 'GROUP', 'AGG_FUNCTION', 'CORR_METHOD', 'TARGET'],
    columns=['METRIC'],
    values='CORR_ROLLBACK_RATIO'
)

# Resetear el índice
table = table.reset_index()

# Ordenar por grupo, nombre de aplicación, método de correlación y target
table = table.sort_values(['GROUP', 'APPLICATION_NAME', 'CORR_METHOD', 'AGG_FUNCTION', 'TARGET'])

# Redondear valores de correlación a 2 decimales
metrics = ['WMC', 'DIT', 'NOC', 'CBO', 'RFC', 'LCOM']
for metric in metrics:
    if metric in table.columns:
        table[metric] = table[metric].round(2)

# Mostrar la tabla
print("\nTabla de correlaciones:")
display(table)

# Guardar la tabla en CSV
table.to_csv("../datasets/correlations/app_correlation_pivoted.csv", index=False)
```
